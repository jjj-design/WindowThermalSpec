### 全体計算例（グレージング複合体のみ）

```python
if __name__ == '__main__':

    #入力値
    #気象データ
    location = '3639999'               #東京ID
    year = 1995                        #西暦年
    lat = 35. + 41.2 / 60.             #計算対象地点の緯度 (°)  
    lon = 139. + 45.9 / 60.            #計算対象地点の経度 (°)  
    Bw = 90.                           #水平面に対する斜面（開口面）の傾斜角 (°) 
    Aw = 0.                            #斜面（開口面）の方位角 (°)
    rho = 0.1                          #アルベド

    #グレージング
    n_gl = 3                             # グレージングの数
    
    g1 = gtb.GlassLayer(gus=[gtb.GlassUnit(d=0.003, lmd=1.0)], sff=gtb.LowESurface(0.837), sfb=gtb.LowESurface(0.837))
    g2 = gtb.GlassLayer(gus=[gtb.GlassUnit(d=0.003, lmd=1.0)], sff=gtb.LowESurface(0.837), sfb=gtb.LowESurface(0.837))
    g3 = gtb.GlassLayer(gus=[gtb.GlassUnit(d=0.003, lmd=1.0)], sff=gtb.LowESurface(0.900), sfb=gtb.LowESurface(0.900))
    
    air = gtb.MixedAirProperty(c_air=100.0, c_argon=0.0, c_sf6=0.0, c_krypton=0.0)
    a1 = gtb.AirLayer(air_property=air, direction=gtb.GlassDirection.VERTICAL, s=0.012)
    a2 = gtb.AirLayer(air_property=air, direction=gtb.GlassDirection.VERTICAL, s=0.012)

    #天空分割数
    Nd_lat = 90                        #緯度方向
    Nd_lon = 180                       #経度方向

    #地表面反射分割数
    Nr_lat = 90                        #緯度方向
    Nr_lon = 180                       #経度方向

    #
    season = 'summer' #夏期：0、冬期：1-->暖冷房期間から決定すべき
    #室温
    ti = {
        'summer': 25.0,
        'winter': 20.0
    }[season]


    gl_in = [
        oip.GlassInput(trs_0_f=0.815, trs_0_b=0.815, ref_0_f=0.072, ref_0_b=0.072, gtype=0, c_type_f=False, c_type_b=False),
        oip.GlassInput(trs_0_f=0.815, trs_0_b=0.815, ref_0_f=0.072, ref_0_b=0.072, gtype=0, c_type_f=False, c_type_b=False),
        oip.RoleInput(trs_0_f=0.3, trs_0_b=0.3, ref_0_f=0.63, ref_0_b=0.63)
    ]
    
    tau_d = TSRTG.actual_trans_diffuse(gl_in, Nd_lat, Nd_lon)
    tau_r = TSRTG.actual_trans_reflect(gl_in, Nr_lat, Nr_lon)

    abs_d = asrig.actual_abs_diffuse(gl_in, Nd_lat, Nd_lon)
    abs_r = asrig.actual_abs_reflect(gl_in, Nr_lat, Nr_lon)

    I_in_J = np.zeros(8760)
    I_tau_J = np.zeros(8760)
    I_qi_J = np.zeros(8760)

    I_abs = np.zeros(n_gl)
    Rglz = np.empty(2 * n_gl + 1, dtype = tuple)
    ghg = np.zeros(8760)

    #気象データ読込
    w = WEA.Weather(location)

    #for i in range(0, 8760):
    for i in range(0, 24):

        nday = int((i + 1.) / 24.) + 1.
        tm = ((i + 1.) / 24. - int((i + 1.) / 24.)) * 24.
        
        sp_h, sp_A = SP.solar_position(year, nday, tm, lat, lon)          # 太陽高度角, 太陽方位角

        dbd_in = SG2BDR.dbd_input(w.rows[i].ig, sp_h)
        dbd_ib = SG2BDR.division_beam_diffuse(dbd_in)[0]                          #法線面直達日射量
        dbd_id = SG2BDR.division_beam_diffuse(dbd_in)[1]                          #水平面天空日射量

        st_in = SROTS.st_input(w.rows[i].ig, dbd_ib, dbd_id, sp_h, sp_A, Bw, Aw, rho)
        st_itb = SROTS.solor_tilt(st_in)[0]                                      #斜面（開口面）直達日射量
        st_itd = SROTS.solor_tilt(st_in)[1]                                      #斜面（開口面）天空日射量
        st_itr = SROTS.solor_tilt(st_in)[2]                                      #斜面（開口面）地表面反射日射量
        phi_b = SROTS.solor_tilt(st_in)[3]                                       #直達日射の入射角

    #単位変換(J2W)
        I_tb = TU.transform_unit_joule2watt(st_itb)                           #斜面（開口面）直達日射量
        I_td = TU.transform_unit_joule2watt(st_itd)                           #斜面（開口面）天空日射量
        I_tr = TU.transform_unit_joule2watt(st_itr)                           #斜面（開口面）地表面反射日射量

    #開口面入射日射量
        ita_in = isrog.ItaInput(I_tb, I_td, I_tr)
        I_in = ita_in.incident_sol()
        I_in_J[i] = TU.transform_unit_watt2joule(I_in)

    #開口面透過日射量
        tau_b = TSRTG.actual_trans_beam(gl_in, phi_b)
        I_tau = tau_b * ita_in.itb + tau_d * ita_in.itd + tau_r * ita_in.itr        
        I_tau_J[i] = TU.transform_unit_watt2joule(I_tau)

    #層jの吸収日射量
        abs_b = asrig.actual_abs_beam(gl_in, phi_b)
        
        I_abs = abs_b * ita_in.itb + abs_d * ita_in.itd + abs_r * ita_in.itr

    #熱平衡
        g = gtb.Glass(gls=[g1, g2, g3], als=[a1, a2])

        # 表面温度
        # 熱抵抗
        # 室内側へ流入する吸収日射量        
        theta, reg, I_qi = g.get_temp_and_r(
            theta_e=w.rows[i].temp,
            theta_i=ti,
            surface_method='JIS_A2103',
            season=season,
            decision_air_layer_temp='calc',
            ia=I_abs
        )
        
    #室内側へ流入する吸収日射量
        I_qi_J[i] = TU.transform_unit_watt2joule(I_qi)

    #時別の日射熱取得率   
        ghg_in = HGOG.ghg_input(I_in_J[i], I_tau_J[i], I_qi_J[i])
        ghg[i] = HGOG.hourly_gvalue_glazing(ghg_in)

        print (i, I_in_J[i], I_tau_J[i], I_qi_J[i], ghg[i])
        #print (i)

    #np.savetxt(location + '_out_I_in.csv', I_in_J, fmt='%.10f', delimiter=',')
    #np.savetxt(location + '_out_I_tau.csv', I_tau_J, fmt='%.10f', delimiter=',')
    #np.savetxt(location + '_out_I_qi.csv', I_qi_J, fmt='%.10f', delimiter=',')
    #np.savetxt(location + '_out_ghg.csv', ghg, fmt='%.10f', delimiter=',')
```
